```python
###' ################################################################################
###'
###' IMPORT LIBRARIES
###'
###'

import numpy as np
import pandas as pd
from pathlib import Path

from keybert import KeyBERT

import ast
import umap.umap_ as umap
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import SpectralEmbedding

######## LDA
```

```python
# Load dataset
data_dir = Path(r"C:\Users\Hyemi\Python\TopicModeling\Data")
data_dir.mkdir(parents=True, exist_ok=True) 

embedding_files = [
    "articles_4_clustering1.csv",
    "articles_4_clustering2.csv",
    "articles_4_clustering3.csv",
    "articles_4_clustering4.csv"
]

embedding_file_paths = [data_dir / file for file in embedding_files]


embedding_dfs = []
for file_path in embedding_file_paths:
    if file_path.exists():  # Check if the file exists before reading
        embedding_dfs.append(pd.read_csv(file_path))

df1a = embedding_dfs[0]
df2a = embedding_dfs[1]
df3a = embedding_dfs[2]
df4a = embedding_dfs[3]
```

## Group 1

```python
group1_df = df1a
group1_df
```

```python
# Initialize KeyBERT model
kw_model = KeyBERT()

# Dictionary to store refined results
keybert_cluster_summaries1 = {}
clusters = group1_df["Cluster3"].unique()

# Iterate through each cluster
for cluster in clusters:
    cluster_data = group1_df[group1_df["Cluster3"] == cluster]
    
    # Combine all abstracts in the cluster
    combined_text = " ".join(cluster_data["Abstract_tokens"])
    
    # Extract meaningful keywords using KeyBERT
    keywords = kw_model.extract_keywords(combined_text, keyphrase_ngram_range=(2,4), stop_words='english', top_n=10)
    extracted_keywords = [kw[0] for kw in keywords]  # Extract just the words
    
    # Generate a summary using the top meaningful keywords
    summary = f"This cluster primarily explores research themes related to {', '.join(extracted_keywords[:5])}, among others."
    
    # Store refined results
    keybert_cluster_summaries1[cluster] = {
        "Keywords": extracted_keywords,
        "Summary": summary
    }
```

```python
keybert_summary_df1 = pd.DataFrame.from_dict(keybert_cluster_summaries1, orient="index")
pd.set_option('display.max_colwidth', None)
keybert_summary_df1
```
