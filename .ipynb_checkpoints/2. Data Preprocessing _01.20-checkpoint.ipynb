## 1. tokenize

```python
def rem_punc_stop(text):
    # When text is None
    if text is None:
        return []

    # Convert text to lowercase
    text = text.lower()

    # Define additional stop words
    stop_words = STOP_WORDS | {"introduction", "objective", "background", "abstract", "purpose", "finding", "find", "result", "study", "aim",  "use",
                               "outcome", "previous", "present", "future", "overall", "test", "effect", "affect", "impact", "suggest",
                               "form", "analysis", "identify", "research", "indicate", "year", "yearold", "age", "know", "high", "low","level"}
                            #   "available", "al", "et", "doi", "googlescholar", "google", "scholar", "textgoogle", 

    # Define punctuation
    punc = set(punctuation)

    # Remove punctuation
    punc_free = "".join([ch for ch in text if ch not in punc])

    # Apply NLP processing
    doc = nlp(punc_free)

    # Tokenize and lemmatize
    text_lemma = " ".join([token.lemma_ for token in doc])

    # Filter tokens to remove URLs, stop words, and non-alphabetic tokens
    filtered_tokens = [word for word in text_lemma.split() if word not in stop_words and word.isalpha()]

    # Return filtered tokens for TfidfVectorizer
    return filtered_tokens
```

```python
articles_full['Abstract_tokens'] = articles_full['Abstract'].map(lambda x: rem_punc_stop(x))
articles_full['Abstract_join'] = articles_full['Abstract_tokens'].map(lambda text: ' '.join(text) if isinstance(text, list) else "")
```

```python
def analyze_multiple_ngram_frequencies(texts, min_df=3, ngram_ranges=[(2,2), (3,3)]):
    
    ngram_frequencies = {}
    
    for ngram_range in ngram_ranges:
        countvec = CountVectorizer(
            min_df=min_df, 
            ngram_range=ngram_range
        )
        
        ngrams = countvec.fit_transform(texts)
        ngram_freq = ngrams.sum(axis=0).A1
        
        df_ngram = pd.DataFrame({
            'ngrams': countvec.get_feature_names_out(),
            'freq': ngram_freq,
            'type': f'{ngram_range[0]}-gram'
        })
        
        ngram_frequencies[f'{ngram_range[0]}-gram'] = df_ngram.sort_values('freq', ascending=False)
    
    return ngram_frequencies

def visualize_ngram_frequencies(ngram_frequencies, top_n=25):
    
    plt.figure(figsize=(20, 10))
    
    for i, (ngram_type, df) in enumerate(ngram_frequencies.items(), 1):
        plt.subplot(1, 2, i)
        top_df = df.head(top_n)
        sns.barplot(x="ngrams", y="freq", data=top_df)
        plt.title(f'Top {top_n} {ngram_type} Frequencies')
        plt.xticks(rotation=90)
        plt.tight_layout()
    
    plt.show()

# Usage example
ngram_frequencies = analyze_multiple_ngram_frequencies(
    articles_full['Abstract_join'], 
    ngram_ranges=[(2,2), (3,3)]
)

# Visualize frequencies
visualize_ngram_frequencies(ngram_frequencies)

# Print top 10 for each n-gram type
for ngram_type, df in ngram_frequencies.items():
    print(f"\nTop 10 {ngram_type}s:")
    print(df.head(10)['ngrams'].values)
```

```python
def analyze_bigram_frequencies_by_year(df, text_column='Abstract_join', year_column='Year', min_df=40):

    # Group unique years
    years = sorted(df[year_column].unique())
    
    # Dictionary to store year-wise bigram frequency data
    year_bigram_frequencies = {}
    total_bigrams_per_year = {}

    for year in years:
        # Filter texts for specific year
        year_texts = df[df[year_column] == year][text_column]
        
        # Skip if no texts for the year
        if len(year_texts) == 0:
            continue
        
        # Create CountVectorizer for bigrams (2-grams)
        countvec = CountVectorizer(min_df=min_df, ngram_range=(2,2))
        
        # Compute n-grams
        bigrams = countvec.fit_transform(year_texts)
        bigram_freq = bigrams.sum(axis=0).A1
        
        # Create DataFrame with bigram frequency
        df_bigram = pd.DataFrame({
            'bigrams': countvec.get_feature_names_out(),
            'freq': bigram_freq,
            'Year': year
        })
        
        # Sort by frequency
        df_bigram = df_bigram.sort_values('freq', ascending=False)
        
        # Store results
        year_bigram_frequencies[year] = df_bigram
        total_bigrams_per_year[year] = bigram_freq.sum()  # Total occurrences of all bigrams in that year
    
    return year_bigram_frequencies, total_bigrams_per_year

# Usage
bigram_frequencies, total_bigrams = analyze_bigram_frequencies_by_year(
    articles_full,
    min_df=40
)

# Print top 5 bigrams for each year and total bigram count per year
for year, df in bigram_frequencies.items():
    print(f"\nYear {year}:")
    print(f"Total bigram occurrences: {total_bigrams[year]}")
    print(df[['bigrams', 'freq']])
```

```python
for year, df_bigram in bigram_frequencies.items():
    # Apply different frequency thresholds based on the year
    if year <= 2020:
        filtered_df = df_bigram[df_bigram['freq'] > 50]
    else:
        filtered_df = df_bigram[df_bigram['freq'] > 100]

    # Print results only if there are bigrams meeting the condition
    if not filtered_df.empty:
        print(f"\nYear {year}:")
        print(f"Total bigram occurrences: {total_bigrams[year]}")
        print(filtered_df[['bigrams', 'freq']])
```
