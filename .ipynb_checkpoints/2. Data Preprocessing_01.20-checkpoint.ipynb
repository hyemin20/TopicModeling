## 1. tokenize

```python
def rem_punc_stop(text):
    # When text is None
    if text is None:
        return []

    # Convert text to lowercase
    text = text.lower()

    # Define additional stop words
    stop_words = STOP_WORDS | {"abstract", "available", "student", "research", "study", "impact", "effect",
                               "result", "al", "et", "doi", "googlescholar", "google", "scholar", "textgoogle", 
                               "full", "crossref", "introduction", "background", "purpose" "aim", "objective","use","child"}

    # Define punctuation
    punc = set(punctuation)

    # Remove punctuation
    punc_free = "".join([ch for ch in text if ch not in punc])

    # Apply NLP processing
    doc = nlp(punc_free)

    # Tokenize and lemmatize
    text_lemma = " ".join([token.lemma_ for token in doc])

    # Filter tokens to remove URLs, stop words, and non-alphabetic tokens
    filtered_tokens = [word for word in text_lemma.split() if word not in stop_words and word.isalpha()]

    # Return filtered tokens for TfidfVectorizer
    return filtered_tokens
```

```python
articles_full['Abstract_tokens'] = articles_full['Abstract'].map(lambda x: rem_punc_stop(x))
articles_full['Abstract_join'] = articles_full['Abstract_tokens'].map(lambda text: ' '.join(text) if isinstance(text, list) else "")
```

```python
def analyze_multiple_ngram_frequencies(texts, min_df=3, ngram_ranges=[(2,2), (3,3)]):
    
    ngram_frequencies = {}
    
    for ngram_range in ngram_ranges:
        countvec = CountVectorizer(
            min_df=min_df, 
            ngram_range=ngram_range
        )
        
        ngrams = countvec.fit_transform(texts)
        ngram_freq = ngrams.sum(axis=0).A1
        
        df_ngram = pd.DataFrame({
            'ngrams': countvec.get_feature_names_out(),
            'freq': ngram_freq,
            'type': f'{ngram_range[0]}-gram'
        })
        
        ngram_frequencies[f'{ngram_range[0]}-gram'] = df_ngram.sort_values('freq', ascending=False)
    
    return ngram_frequencies

def visualize_ngram_frequencies(ngram_frequencies, top_n=25):
    
    plt.figure(figsize=(20, 10))
    
    for i, (ngram_type, df) in enumerate(ngram_frequencies.items(), 1):
        plt.subplot(1, 2, i)
        top_df = df.head(top_n)
        sns.barplot(x="ngrams", y="freq", data=top_df)
        plt.title(f'Top {top_n} {ngram_type} Frequencies')
        plt.xticks(rotation=90)
        plt.tight_layout()
    
    plt.show()

# Usage example
ngram_frequencies = analyze_multiple_ngram_frequencies(
    articles_full['Abstract_join'], 
    ngram_ranges=[(2,2), (3,3)]
)

# Visualize frequencies
visualize_ngram_frequencies(ngram_frequencies)

# Print top 10 for each n-gram type
for ngram_type, df in ngram_frequencies.items():
    print(f"\nTop 10 {ngram_type}s:")
    print(df.head(10)['ngrams'].values)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_ngram_frequencies_by_year(df, text_column='Abstract_join', year_column='Year', 
                                       min_df=3, ngram_ranges=[(2,2), (3,3)]):
    """
    Analyze n-gram frequencies across different years
    
    Args:
        df (pd.DataFrame): DataFrame with text and year columns
        text_column (str): Name of column containing text
        year_column (str): Name of column containing publication year
        min_df (int): Minimum document frequency
        ngram_ranges (list): List of n-gram range tuples
    
    Returns:
        dict: N-gram frequencies by year
    """
    # Group unique years
    years = sorted(df[year_column].unique())
    
    # Initialize results dictionary
    year_ngram_frequencies = {}
    
    for ngram_range in ngram_ranges:
        # Prepare year-wise n-gram frequency tracking
        year_frequencies = {}
        
        for year in years:
            # Filter texts for specific year
            year_texts = df[df[year_column] == year][text_column]
            
            # Skip if no texts for the year
            if len(year_texts) == 0:
                continue
            
            # Create CountVectorizer
            countvec = CountVectorizer(
                min_df=min_df, 
                ngram_range=ngram_range
            )
            
            # Compute n-grams
            ngrams = countvec.fit_transform(year_texts)
            ngram_freq = ngrams.sum(axis=0).A1
            
            # Create DataFrame
            df_ngram = pd.DataFrame({
                'ngrams': countvec.get_feature_names_out(),
                'freq': ngram_freq,
                'Year': year
            })
            
            # Sort and store
            year_frequencies[year] = df_ngram.sort_values('freq', ascending=False)
        
        year_ngram_frequencies[f'{ngram_range[0]}-gram'] = year_frequencies
    
    return year_ngram_frequencies

# Usage
year_ngram_frequencies = analyze_ngram_frequencies_by_year(
    articles_full, 
    ngram_ranges=[(2,2), (3,3)]
)

# Print top 5 n-grams for each year and n-gram type
for ngram_type, year_data in year_ngram_frequencies.items():
    print(f"\n{ngram_type.upper()} Frequencies by Year:")
    for year, df in year_data.items():
        print(f"\nYear {year}:")
        print(df.head(5)['ngrams'].values)
```
